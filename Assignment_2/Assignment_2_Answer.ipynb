# 4 - Visulaization

## Output
The output is 'price' coloum

```python
import matplotlib.pyplot as plt

# Extract the 'price' column
prices = car_prices_data['price']

# Create a histogram to visualize the distribution of prices using Matplotlib
plt.figure(figsize=(8, 6))
plt.hist(prices, bins=20, edgecolor='k')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## Quantitvative Data 

Use histogram or any other preferred method you prefer to visualize your data (one feature is enough)

```python
# Visualize the 'enginesize' feature with a histogram
plt.figure(figsize=(8, 6))
plt.hist(car_prices_data['enginesize'], bins=15, color='red', edgecolor='black')
plt.title('enginesize Distribution')
plt.xlabel('enginesize')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Visualize the Correlation Matrix

```python
import seaborn as sns

# Select the quantitative features for the correlation matrix
quantitative_features = car_prices_data.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
correlation_matrix = quantitative_features.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Select the 'engine-size' feature
engine_size = car_prices_data['enginesize']

# Create a histogram 
plt.figure(figsize=(8, 6))
plt.hist(engine_size, bins=20, edgecolor='k')
plt.title('Engine Size Distribution')
plt.xlabel('Engine Size')
plt.ylabel('Frequency')
plt.show()



# Check for missing data 
missing_data = car_prices_data['enginesize'].isnull().sum()
print(f"Missing Data in 'engine-size': {missing_data} values")


# Calculate the correlation between 'engine-size' and 'price'
correlation = car_prices_data['enginesize'].corr(car_prices_data['price'])
print(f"Correlation between 'engineSize' and 'price': {correlation:.2f}")
```

## Qualitative Data 

Use boxplot or any other preferred method you prefer to visualize your data (one feature is enough)

```python
# Count the frequency of each 'car_body' category
car_body_counts = car_prices_data['carbody'].value_counts()

# Create a pie chart 
plt.figure(figsize=(8, 8))
plt.pie(car_body_counts, labels=car_body_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Car Bodies')
plt.show()
```
